I have also checked how the model is performing on the test data: 

```python
y_pred = lg1.predict(X_test)
```

```python
cm = metrics.confusion_matrix(Y_test, y_pred) #There are 1500 points in test data. 
```

```python
cm
```

```python
TP = cm[1,1]
TN = cm[0,0]
FP = cm[0,1]
FN = cm[1,0]
```

```python
print(TP,TN,FP,FN)
```

As we can see that the FN is quite high in this case which could be an issue fo

```python
from sklearn.metrics import accuracy_score,f1_score,precision_score,recall_score, roc_curve, auc, roc_auc_score,confusion_matrix,classification_report
```

```python
accuracy = (TP + TN)/(TP + FP + FN + TN)
recall_TPR = (TP)/ (TP + FN)
precision = (TP)/ (TP + FP)
roc_auc = roc_auc_score(Y_test, y_pred)
print('accuracy is :', accuracy)
print('recall_TPR is :', recall_TPR)
print('precision is :', precision)
print('RoC score is :', roc_auc)
```

```python
report = metrics.classification_report(y_pred, Y_test)
print(report)
```

```python
reciver_curve = roc_curve(y_pred, Y_test)
```

```python
false_positive_rate, true_positive_rate, thresholds = roc_curve(Y_test, y_pred)
plt.title('Receiver Operating Characteristic')
plt.plot(false_positive_rate, true_positive_rate)
plt.plot([0,1],[0,1],'r--') 
plt.ylabel('True Positive Rate(Sensitivity)')
plt.xlabel('False Positive Rate(1-Specificity)')
plt.show()

print("Area under the curve: ",round(auc(false_positive_rate, true_positive_rate),3))
```

As we can see here the model has 78.5% area under the curve which is quite good for the unseen data. Although the accuracy can be improved if more feautres are examined and taken into the consideration for this analysis. There is a trade off b/w the Specificity and Sensitivity as we have to check which one is more important to us. 
Sensitivity = (TP)/(TP + FN) and (1 - Specificity)= (FP)/(TN + FP) also known as False positive rate.
